### 🧪 Lab: Governance, Risk, and Compliance (GRC) in a SOC Analyst \\Agentic **AI**

In this lab, we explore how **Governance, Risk, and Compliance (GRC)** apply to an
MCP-powered SOC Analyst AI agent. We simulate how **unfiltered MCP exchanges** can cause
compliance failures (e.g., GDPR/HIPAA violations), and how applying **governance controls**
(anonymization, access control, and logging) fixes the issue.

We follow the **Build → Break → Defend** pattern:

1. **Build:** SOC Analyst AI processes security alerts using MCP.  
2. **Break:** Raw context includes PII (usernames, IPs) → compliance failure.  
3. **Defend:** Apply GRC filters → anonymization, access control, logging → compliance success.  



## Part 1: Build (SOC Analyst AI with MCP)

We simulate a SOC Analyst AI agent that processes alerts with the **Model Context Protocol (MCP)**.  
The alert includes sensitive data such as usernames and IP addresses.  


```python

import json

# Sample alert with sensitive data (PII)
alert = {
    "alert_id": "SOC12345",
    "username": "jdoe",
    "ip_address": "192.168.1.100",
    "event": "Failed login attempt",
    "severity": "High"
}

# MCP context creation
def mcp_create_context(user, alert):
    context = {
        "analyst": user,
        "alert": alert
    }
    return json.dumps(context)

def mcp_exchange_context(context_str):
    return json.loads(context_str)

# SOC Analyst AI agent (no compliance filters)
def soc_agent_process(context):
    alert = context["alert"]
    return f"[SOC Agent] Processed alert for {alert['username']} from {alert['ip_address']} - {alert['event']}"

# Example flow
context = mcp_create_context("SOC_AI", alert)
parsed_context = mcp_exchange_context(context)
soc_agent_process(parsed_context)
```


## Part 2: Break (Compliance Failure)

Because the SOC Analyst AI logs **raw usernames and IPs**, it violates GDPR/HIPAA by exposing PII.  
This would fail an audit or regulatory review.  


```python

# Raw log (non-compliant)
log_entry = soc_agent_process(parsed_context)
print("Raw Log Entry:", log_entry)
print("⚠️ Compliance Failure: PII exposed in logs")
```


## Part 3: Defend (Apply GRC Controls)

We add **compliance filters** to MCP and the SOC agent:

1. **Anonymization:** Mask usernames and IPs.  
2. **Access Control:** Only authorized roles can view raw PII.  
3. **Audit Logging:** Maintain secure logs of compliance decisions.  


```python

import hashlib

AUTHORIZED_ROLES = ["compliance_officer"]

def anonymize_alert(alert):
    """Mask sensitive fields (username, IP)."""
    masked_alert = alert.copy()
    masked_alert["username"] = "user_" + hashlib.sha256(alert["username"].encode()).hexdigest()[:6]
    masked_alert["ip_address"] = "xxx.xxx.xxx.xxx"
    return masked_alert

def mcp_secure_exchange(user_role, context_str):
    context = json.loads(context_str)
    alert = context["alert"]

    if user_role not in AUTHORIZED_ROLES:
        # Apply anonymization for normal roles
        context["alert"] = anonymize_alert(alert)
    return context

# Secure flow with anonymization
secure_context = mcp_secure_exchange("soc_analyst", context)
secure_result = soc_agent_process(secure_context)
print("Secure Log Entry:", secure_result)
print("✅ Compliance Success: PII anonymized for standard roles")
```
